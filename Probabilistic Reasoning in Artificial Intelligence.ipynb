### 2. Rolling Dice (12 points)

Consider two die with 6 equally weighted sides.

### a. What is the probability of a dice roll summing to 9?

The pairs of rolls that sum to 9 are: (3,6), (4,5), (5,4), (6,3)

$P(\text{Sum} = 9) = \frac{4}{36} = \frac{1}{9}$

### b. What is the probability of a dice roll summing to an odd number 8 times in a row?

A sum is odd when one die is even and the other is odd. Each die has **3 even numbers** (2, 4, 6) and **3 odd numbers** (1, 3, 5). 

The probability of rolling an odd sum in one roll is:

$P(\text{Odd Sum}) = P(\text{Odd, Even}) + P(\text{Even, Odd}) = \left(\frac{3}{6} \times \frac{3}{6} \right) + \left(\frac{3}{6} \times \frac{3}{6} \right) = \frac{9}{36} + \frac{9}{36} = \frac{18}{36} = \frac{1}{2}$

Since each roll is independent, the probability of rolling an odd sum 8 times in a row is:

$P(\text{8 Odd Sums}) = \left(\frac{1}{2}\right)^8 = \frac{1}{256}$

### c. Let's say that you roll one dice at a time. Your first die comes up as an odd number. What is the probability that your dice will sum to 9 after you roll the second die?

The sum must be 9. If the first die is odd, the second die must be **even** for the total to be 9.

Possible first die values: **1, 3, 5**  
For each odd value, the second die must be:  
- **If first die is 1**, second die must be 8 (not possible) ❌  
- **If first die is 3**, second die must be 6 ✅  
- **If first die is 5**, second die must be 4 ✅  

There are **2 successful outcomes** out of **3 possible odd first rolls**. The probability of rolling the required even number on the second die is:

$P(\text{Sum} = 9 | \text{First Die is Odd}) = \frac{2}{6} = \frac{1}{3}$

### 4. Bayes' Rule (20 points)

### (a). 
Two bags contain different combinations of candies. Bag 1 has six chocolates and four gummy bears, while Bag 2 has five chocolates and five gummy bears. A bag is chosen at random: P(bag=1) = P(bag=2) = 0.5, and a candy drawn from this bag happens to be a chocolate. What is the probability that the chocolate came from Bag 1? (i.e., P(bag=1 | candy=chocolate)=?)

Two bags contain different combinations of candies. Bag 1 has six chocolates and four gummy bears, while Bag 2 has five chocolates and five gummy bears. A bag is chosen at random: 

$P(\text{Bag 1}) = P(\text{Bag 2}) = 0.5$

A candy drawn from this bag happens to be a chocolate. What is the probability that the chocolate came from Bag 1?

$P(\text{Bag 1} | \text{Chocolate}) = \frac{P(\text{Chocolate} | \text{Bag 1}) P(\text{Bag 1})}{P(\text{Chocolate})}$

Where:
- \( $P(\text{Chocolate} | \text{Bag 1}) = \frac{6}{10} = 0.6$ \)
- \( $P(\text{Chocolate} | \text{Bag 2}) = \frac{5}{10} = 0.5$ \)
- By **Law of Total Probability**:

$P(\text{Chocolate}) = P(\text{Chocolate} | \text{Bag 1}) P(\text{Bag 1}) + P(\text{Chocolate} | \text{Bag 2}) P(\text{Bag 2})$

$P(\text{Chocolate}) = (0.6 \times 0.5) + (0.5 \times 0.5) = 0.3 + 0.25 = 0.55$

$P(\text{Bag 1} | \text{Chocolate}) = \frac{0.6 \times 0.5}{0.55} = \frac{0.3}{0.55} \approx 0.5455$

### (b).

Two cards are placed in an envelope as follows: A fair six-sided die is rolled, and a blue card is placed in the envelope if an even number (2, 4, or 6) comes up, otherwise a red card is placed in the envelope. The die is rolled again, and a red card is placed in the envelope if an odd number (1, 3, or 5) comes up, otherwise a blue card is placed in the envelope. Cards are drawn from the envelope three times in succession (always replacing the drawn card back in the envelope). It is found that on all three occasions a red card is drawn. What is the probability that both cards in the envelope are red?

Define:
- \( $P(R) = \frac{3}{6} = 0.5 $\) (Probability of rolling an odd number → Red card)
- \( $P(B) = \frac{3}{6} = 0.5 $\) (Probability of rolling an even number → Blue card)

Possible cases:
1. **Both Red (RR)**: Probability \( $P(RR) = P(R) \times P(R) = 0.5 \times 0.5 = 0.25$ \)
2. **One Red, One Blue (RB or BR)**: Probability \( $P(RB) = P(R) \times P(B) = 0.5 \times 0.5 = 0.25$ \)
3. **Both Blue (BB)**: Probability \( $P(BB) = P(B) \times P(B) = 0.5 \times 0.5 = 0.25$ \)

$P(RR | \text{3 Reds Drawn}) = \frac{P(\text{3 Reds Drawn} | RR) P(RR)}{P(\text{3 Reds Drawn})}$

Where:
- \( $P(\text{3 Reds Drawn} | RR) = 1$ \)
- \( $P(\text{3 Reds Drawn} | RB) = P(\text{Drawing Red from RB})^3 = \left(\frac{1}{2}\right)^3 = \frac{1}{8}$ \)
- \( $P(\text{3 Reds Drawn} | BB) = 0$ \)

$P(\text{3 Reds Drawn}) = (1 \times 0.25) + \left(\frac{1}{8} \times 0.5\right) = 0.25 + 0.0625 = 0.3125$

$P(RR | \text{3 Reds Drawn}) = \frac{1 \times 0.25}{0.3125} = \frac{0.25}{0.3125} = 0.8$

### (c).
A secret government agency has developed a scanner that determines whether a person is an alien imposter. The scanner detects alien imposters with 95% reliability. However, 1% of the time the test returns a false positive for an upstanding citizen. The alien imposters are very dangerous, but also quite rare, with a prevalence in the population of only 0.1%. Upon a tip from an informant, the agency has administered the scanner to a suspicious person. 

#### i. 
Use the *Law of Total Probability* to calculate the probability that the suspicious person will test positive.

#### ii.
What is the probability that the person is actually an alien imposter given that they do test positive?

Define:
- $( P(A) = 0.001 )$ (Probability that a person is an **Alien Imposter**)
- $( P(H) = 0.999 )$ (Probability that a person is **Human**)
- $( P(\text{Test Positive} | A) = 0.95 )$ (True Positive Rate)
- $( P(\text{Test Positive} | H) = 0.01 )$ (False Positive Rate)

Using **Law of Total Probability**:

$P(\text{Test Positive}) = P(\text{Test Positive} | A) P(A) + P(\text{Test Positive} | H) P(H)$

$P(\text{Test Positive}) = (0.95 \times 0.001) + (0.01 \times 0.999)$

$P(\text{Test Positive}) = 0.00095 + 0.00999 = 0.01094$

#### (ii) What is the probability that the person is actually an **alien imposter** given that they do test positive?

$P(A | \text{Test Positive}) = \frac{P(\text{Test Positive} | A) P(A)}{P(\text{Test Positive})}$

$P(A | \text{Test Positive}) = \frac{0.95 \times 0.001}{0.01094}$

$P(A | \text{Test Positive}) = \frac{0.00095}{0.01094} \approx 0.0868$
